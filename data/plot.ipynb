```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
kaanan_csv = pd.read_csv("/Users/klau/Projects/loci-map/loci-visualize/data/kaanan.csv")
erik_csv = pd.read_csv("/Users/klau/Projects/loci-map/loci-visualize/data/erik.csv")
aidan_csv = pd.read_csv("/Users/klau/Projects/loci-map/loci-visualize/data/aidan.csv")
kenny_csv = pd.read_csv("/Users/klau/Projects/loci-map/loci-visualize/data/kenny.csv")
kenny_csv = kenny_csv[kenny_csv["Conductance"] < 1]
erik_csv = erik_csv[erik_csv["Conductance"] < 1]
aidan_csv = aidan_csv[aidan_csv["Conductance"] < 1]
kaanan_csv = kaanan_csv[kaanan_csv["Conductance"] < 1]
```

```python
kenny_csv["Seconds"] = kenny_csv["Timestamp"] - kenny_csv["Timestamp"].min()
kenny_csv["Seconds Rounded"] = round(kenny_csv["Seconds"])
kenny_csv = kenny_csv.groupby("Seconds Rounded").mean().reset_index()
plt.plot(kenny_csv["Seconds Rounded"], kenny_csv["Conductance"])
plt.xlabel("Seconds")
plt.ylabel("Conductance")
plt.show()
```
